```python
import pandas as pd
from google.colab import files


#uploaded_files = files.upload()

# load the datasets into pandas df
movie_df = pd.read_csv('movie_dataset.csv')
nupur1_df = pd.read_csv('nupur1_db.csv')

# Combine columns with the same name
combined_df = pd.concat([movie_df, nupur1_df], axis=0, ignore_index=True)

# Merge similar columns
# Fill NaNs to handle missing data before merging
combined_df['merged_keywords'] = combined_df['keywords'].fillna('') + ' ' + combined_df['cleaned_overview'].fillna('')

# drop redundancy
combined_df.drop(['keywords', 'cleaned_overview'], axis=1, inplace=True)

# save new df to a csv file
output_filename = 'combined_dataset.csv'
combined_df.to_csv(output_filename, index=False)

#download new csv file
files.download(output_filename)
```

```python
combined_df['combined_genres'] = combined_df['genres'].fillna('') + ' ' + combined_df['genre'].fillna('')
```

```python
combined_df['combined_title'] = combined_df['title'].fillna('') + ' ' + combined_df['movie_name'].fillna('')
```

```python
combined_df.drop(['genres', 'genre', 'title', 'movie_name'], axis=1, inplace=True)
```

```python
output_filename = 'nupur2_db.csv'
combined_df.to_csv(output_filename, index=False)

# Allow user to download the file
files.download(output_filename)
```
